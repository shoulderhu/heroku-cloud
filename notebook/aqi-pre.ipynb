```python
def csv_modify(file, loc, zone):
    a = pd.read_csv(f"csv/{file}.csv")
    b = a.drop(["rh", "temp", "rain", "no", "nox", "wd", "ws", "preasure"], axis=1)
    b["測站"] = loc
    b["空品區"] = zone
    b.to_csv(f"{file}.csv", index=False)
```

```python
def csv_modify2(file, loc, zone, header=False):
    a = pd.read_csv(f"csv/{file}.csv")
    b = a.drop(["month", "week", "hr"], axis=1)
    
    t = datetime(2019, 1, 1)
    times = []
    for i in range(2880):
        times.append(str(t))
        t = t + timedelta(hours=1)
        
    b.insert(8, "datetime", times)    
    b.to_csv(f"csv2/{file}.csv", index=False, header=False)
```
